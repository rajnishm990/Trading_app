```python
import helpers.clients as helpers
```

```python
company_name = "Google"
company_ticker = "GOOG"
multiplier = 1
from_date = "2024-01-09"
to_date = "2024-05-09"
```

```python
client = helpers.PolygonAPIClient(
    ticker=company_ticker,
    multiplier=multiplier,
    from_date=from_date,
    to_date=to_date
)
dataset = client.get_stock_data()
```

```python
len(dataset)
```

```python
from market.models import Company, StockQuote
```

```python
obj ,created = Company.objects.get_or_create(name=company_name ,  ticker=company_ticker)
```

```python
obj.name
```

```python
#Bulk updation
stock_list = []
for data in dataset:
    stock_list.append(StockQuote(company=obj, **data))
    
```

```python
len(stock_list)
```

```python
print(stock_list)
```

```python
StockQuote.objects.bulk_create(stock_list,ignore_conflicts=True)
```
